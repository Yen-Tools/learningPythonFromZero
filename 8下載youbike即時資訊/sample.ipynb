```python
#下載所有資訊
import requests
url = 'https://tcgbusfs.blob.core.windows.net/blobyoubike/YouBikeTP.json'
response = requests.get(url)
if response.status_code == 200:
    json=response.json()
    print(json)
else:
    print("下載失敗")
```

```python
allData = list(json["retVal"].values())
allData
```

```python
print("{:<30s}\t\t{:<10s}{:>5s}{:>5s}{:>5s}".format("站名","區域","可借數量","可還數量","總數量"))
for site in allData:
    siteName = site["sna"]
    area = site["sarea"]
    sbi = site["sbi"]
    bemp = site["bemp"]
    total = site["tot"]
    print("{:<30s}\t{:<10s}{:>10s}{:>10s}{:>10s}".format(siteName,area,sbi,bemp,total))
```
