# 台灣銀行匯率下載網址(https://rate.bot.com.tw/xrt?Lang=zh-TW)
# 下載csv至專案中

```python
import csv
import pprint

with open('ExchangeRate.csv',encoding='utf-8',newline='') as file:
    csvReader = csv.reader(file)
    rates = list(csvReader)

#assert False,"輸入有誤"

pp = pprint.PrettyPrinter(
    indent=1,
    width=200,
    compact=True
)
pp.pprint(rates)
```
